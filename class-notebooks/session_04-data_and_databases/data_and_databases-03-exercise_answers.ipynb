        SELECT AVG( exityr - curadmyr ) AS average_time_spent
        FROM ildoc_exit
        WHERE admtypr = 'PVN'
            AND exityr > 1994;

Questions:

- 5) Based on information in ildoc_exit, how many individuals were released from IL correctional facilities in 2013?

        SELECT COUNT( DISTINCT docnbr ) AS number_released
        FROM ildoc_exit
        WHERE exityr = 2013;

- 6) How many individuals were released for each year from 2010-2015?

        SELECT exityr, count( DISTINCT docnbr ) AS number_released
        FROM ildoc_exit
        WHERE exityr >= 2010 AND exityr <= 2015
        GROUP BY exityr;

- 7) How long was the average individual who was released in 2014 in prison for each revised admission type?

        SELECT admtypr, AVG( exityr - curadmyr ) AS average_stay
        FROM ildoc_exit
        WHERE exityr = 2014
        GROUP BY admtypr;

        SELECT docnbr, COUNT( * ) as admit_count
        FROM ildoc_admit
        GROUP BY docnbr
